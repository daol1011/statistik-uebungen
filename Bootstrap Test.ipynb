```python
import pandas as pd
import plotly.express as px

pd.options.display.max_rows = 70
```

```python
cases = pd.DataFrame({"Ergebnis": 3*["Komplikation"] + 59*["Keine Komplikation"]})
```

```python
cases
```

```python
n_boot = 10000
n_comp = []
for k in range(n_boot):
    n_comp.append((cases.sample(62, replace=True)["Ergebnis"] == "Komplikation").sum()/62)
```

```python
px.histogram(n_comp)
```
