# Einführung in Jupyter Notebook

### Was sind Jupyter Notebooks?

* Jupyter Notebook ist eine interaktive web-basierte Entwicklungsumgebung für die Programmiersprachen Python, Julia und R.

* Jupyter Notebooks erlauben es, ausführbaren Programmcode, Visualisierung, Text und in begrenztem Umfang auch Daten in einem Dokument zu speichern,  darzustellen und interaktiv zu bearbeiten.

* Das Dateiformat hat hierbei die Endung .ipynb.

Eine detailliertere Einführung sowie die offizielle Dokumentation findet sich unter  
https://jupyter-notebook.readthedocs.io/en/stable/notebook.html.

### Das Zellen-System

Jupyter Notebooks sind aus einzelnen Zellen aufgebaut, die untereinander angezeigt werden. Es gibt hierbei drei verschiedene Zelltypen:
 1. Code (ausführbarer Programcode - für diese Vorlesung nutzen wir hierfür Python)
 2. Markdown (Beschreibung als Text)
 3. Raw NB-convert (für die Vorlesung nicht benötigt)
 
Die bisherige Beschreibung besteht aus Markdown-Zellen. So sieht eine Code-Zelle aus:

```python
x=7
print(x+1)
```

Code und Markdown-Zellen können sich in beliebiger Weise abwechseln, so dass eine Datenanalyse gut dokumentiert und visualisiert werden kann. Codezellen können zusätzlich auch Grafiken erzeugen:

```python
import matplotlib.pyplot as plt
import numpy as np
import math

X=np.linspace(0,20,100)
Y=np.sin(X)
plt.figure(figsize=(10,3))
plt.plot(X,Y);
```

Eine Code-Zelle kann durch den Shortcut Ctrl-Enter ausgeführt werden.  
Markdown-Zellen können durch den Shortcut Ctrl-Enter gerendert werden.

### Modes und die Tastatureingabe

Die Benutzeroberfläche befindet sich stets entweder im **Command-Mode** oder im **Enter Mode**.

**Command-Mode aktivieren: Esc.**  
Dann sind beispielsweise folgende Befehle verfügbar:
* A: Neue Zelle über aktueller Zelle einfügen (above)
* B: Neue Zelle unter aktueller Zelle einfügen (below)
* M: Zelle in Markdown-Zelle konvertieren.
* X: Zelle löschen

**Enter-Mode aktivieren: Enter**.  
Dann können Markdown- bzw Code-Zellen bearbeitet werden. Außerdem existieren einige Keyboard-shortcuts. Eine Übersicht lässt sich im Command-Mode über den Befehl H anzeigen.

### Formatierung von Markdown-Zellen:

Markdown-Zellen erlauben es Beschreibungen, Erklärungen, Dokus, ... als Text darzustellen. Markdown selbst ist eine auf HTML-basierende Markup-Sprache. Eine ausführliche Dokumentation findet sich unter 

https://daringfireball.net/projects/markdown/

Die wichtigsten Formatierungsmöglichkeiten sind

#### Aufzählungen mittels *

\* Punkt 1  
\* Punkt 2   
\* Punkt 3 


#### Überschriften mittels Raute

\# Überschrift 1    
\## Überschrift 2    
\### Überschrift 3   
\#### Überschrift 4    

#### Fettdruck mittels zwei Sternchen
\*\*beispieltext\*\*

### Interaktivität durch IPython

Jupyter nutzt **IPython** (https://ipython.org/) als sogenannten Kernel. Dies ist eine Shell, die es erlaubt interaktiv mit dem Code zu arbeiten. Sie können deswegen zu jedem Zeitpunkt einzelne Code-Zellen über den Shortcut Ctrl-Enter ausführen.

Ein Notebook startet mit einer neuen, leeren IPython-Session. Das Ergebnis von jeder folgenden Code-Ausführung wird hierbei "gemerkt". wir können eine Liste mit allen Objekten, die aktuell in einer Session existieren über den Befehl dir() aufrufen.

```python
namen=["Egon", "Hans", "Otto"]
```

```python
dir()
```

```python
print(namen)
```

```python
namen=namen + ['Karl']
```

```python
print(namen)
```

Die Zahl in eckigen Klammern links neben jeder Codezeile gibt die Reihenfolge an, in der die Codezeilen zuletzt ausgeführt wurden. Wurde eine Zeile noch nicht ausgeführt, so sind die Klammern leer. Dauert die Ausführung einer Zeile sehr lange, befindet sich während der Ausführung ein * zwischen den Klammern.
